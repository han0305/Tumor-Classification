```python
#import required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#getting data
data = pd.read_csv('data.csv')
```

```python
#data
data.head()
```

```python
# 'Unnamed: 32' consists of only missiong values
data['Unnamed: 32'].isnull().mean()
```

```python
# 'id' consists of all unique values
print(len(data))
print(len(data['id'].unique()))
```

```python
# only 'Unnamed: 32' contains missing value
data.isnull().mean()
```

```python
# description of data
data.describe()
```

```python
#remove id,unamed:32 and diagnosis(target variable)
drop_list = ['id','diagnosis','Unnamed: 32']
target = 'diagnosis'
x = data.drop(drop_list,axis = 1)
y = data[target]
```

```python
x.head()
```

```python
y.head()
```

```python
# checking for outliers
for var in x.columns:
    
    plt.figure(figsize=(15,6))
    plt.subplot(1, 2, 1)
    fig = x.boxplot(column=var)
    fig.set_title('')
    fig.set_ylabel(var)
    
    plt.subplot(1, 2, 2)
    fig = x[var].hist(bins=20)
    fig.set_ylabel('Number')
    fig.set_xlabel(var)
 
    plt.show()
    
```

```python
# skewed data
skewed_list = ['perimeter_mean','area_mean','compactness_mean','concavity_mean','concave points_mean','fractal_dimension_mean','radius_se','texture_se','perimeter_se','area_se','smoothness_se','concavity_se','fractal_dimension_se','radius_worst','perimeter_worst','area_worst','compactness_worst','concavity_worst']
gaussian_list  = [col for col in x.columns if col not in skewed_list]
```

```python
print("skewed list-> {}".format(skewed_list))
print("gaussin list-> {}".format(gaussian_list))
```

```python
print(len(skewed_list)+len(gaussian_list))
print(x.shape)
```

```python
#topcoding
cols = x.columns
for col in cols:
    if col in skewed_list:
        IQR = x[col].quantile(0.75) - x[col].quantile(0.25)
        Upper_fence = data[col].quantile(0.75) + (IQR * 3)
        x[col]=np.where(x[col]>Upper_fence,Upper_fence,x[col])
    else:
        Upper_boundary = data[col].mean() + 3* data[col].std()
        x[col]=np.where(x[col]>Upper_boundary,Upper_boundary,x[col])
```

```python
x.describe()
```

```python
#correlation among variables
f,ax = plt.subplots(figsize=(18, 18))
sns.heatmap(x.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

```python
drop_list = ['perimeter_mean','radius_mean','compactness_mean','concave points_mean','radius_se','perimeter_se','radius_worst','perimeter_worst','compactness_worst','concave points_worst','compactness_se','concave points_se','texture_worst','area_worst']
x.drop(drop_list,inplace=True,axis = 1 ) 
x.head()
```

```python
f,ax = plt.subplots(figsize=(18,18))
sns.heatmap(x.corr(), annot=True, linewidths=.5, fmt= '.1f',ax=ax)
```

```python
#splitting the data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

```python
# importing packages
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
clf_rf = RandomForestClassifier(random_state=43)      
clr_rf = clf_rf.fit(x_train,y_train)
```

```python
# feature selection
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
select_feature = SelectKBest(chi2, k=5).fit(x_train, y_train)
```

```python
print('Score list:', select_feature.scores_)
print('Feature list:', x_train.columns)
```

```python
x_train_2 = select_feature.transform(x_train)
x_test_2 = select_feature.transform(x_test)

clf_rf_2 = RandomForestClassifier(n_estimators=10,criterion='gini')      
clr_rf_2 = clf_rf_2.fit(x_train_2,y_train)
ac_2 = accuracy_score(y_test,clf_rf_2.predict(x_test_2))
print('Accuracy is: ',ac_2)
cm_2 = confusion_matrix(y_test,clf_rf_2.predict(x_test_2))
sns.heatmap(cm_2,annot=True,fmt="d")
```

```python
#hyperparameter tuning
from sklearn.model_selection import GridSearchCV
```

```python
parameters = [{'n_estimators':[10,100,1000],'criterion':['gini']},{'n_estimators':[10,100,1000],'criterion':['entropy']}]
```

```python
grid_search = GridSearchCV(estimator = clr_rf_2, param_grid = parameters, scoring = 'accuracy', cv=10)
```

```python
grid_search = grid_search.fit(x_train_2,y_train)
```

```python
best_score = grid_search.best_score_
```
